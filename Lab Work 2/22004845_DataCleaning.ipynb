```python
import pandas as pd
import numpy as np
```

```python
white_data = pd.read_csv('winequality-white.csv', sep=';')
print(white_data.shape)
white_data['is_red'] = 0
white_data.head()
```

```python
white_data = white_data.drop_duplicates()
print(white_data.shape)
white_data.head()
```

```python
red_data = pd.read_csv('winequality-red.csv', sep=';')
print(red_data.shape)
red_data['is_red'] = 1
red_data.head()
```

```python
red_data = red_data.drop_duplicates()
print(red_data.shape)
red_data.head()
```

```python
data = pd.concat([red_data, white_data], ignore_index=True)
print(data.shape)
```

```python
from sklearn.utils import shuffle

# Shuffle the combined dataset
data = shuffle(data, random_state=42)
data.head()
```

```python
data.tail()
```

```python
data.to_csv('fullwinequality.csv', index = False)
```
